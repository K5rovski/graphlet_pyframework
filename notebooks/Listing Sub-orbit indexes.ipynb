```python
"Listing Sub-orbit indexes"
import sys
import itertools

sys.path.append('..')
from multiplex_pygraph.Multiplex_Graph import GraphFunc


# ==========================================
# ==========================================
edge_types='1;2'.split(';')
# ==========================================
# ==========================================

n=len(edge_types)
Cn=(2**(n))-1
all_edgetypes=[]

for c_n in range(1,n+1):
    for combs in itertools.combinations(edge_types,c_n):
        all_edgetypes.append(';'.join(combs))



graphF=GraphFunc(doDirected=False)
graphF.edgeN=Cn

doSummaryInd=True

sumInd=0

print('orbit 0\n====\n')
for i,e in enumerate(all_edgetypes):
    print("a: '{}' edgetypes of neighbor, yields index: {}".format(e,i ))

sumInd+=i+1
print('\n\norbit 1\n======\n')
for ind,(a,b) in enumerate(itertools.product(range(Cn),range(Cn))):
    print("a: '{}', b: '{}' edgetypes of wedgepath, yields index: {}"\
          .format(all_edgetypes[a],all_edgetypes[b],(a*Cn)+b  + (sumInd if doSummaryInd else 0) ))

sumInd+=ind+1
print('\n\norbit 2\n=====\n')
for ind,(a,b) in enumerate(itertools.combinations_with_replacement(range(Cn),2)):
    print("a: '{}', b: '{}' edgetypes of wedgestar, yields index: {}"\
          .format(all_edgetypes[a],all_edgetypes[b],graphF.calc_combR2(a,b)  + (sumInd if doSummaryInd else 0) ))

sumInd+=ind+1
print('\n\norbit 3\n====\n')
for c,(a,b) in itertools.product(range(Cn),\
                itertools.combinations_with_replacement(range(Cn),2)):
    print("a: '{}', b: '{}', c: '{}', edgetypes of triangle, yields index: {}"\
          .format(all_edgetypes[a],all_edgetypes[b],
                  all_edgetypes[c],graphF.calc_combR2(a,b) +(c*graphF.comb2_num()) + (sumInd if doSummaryInd else 0)  ))
```
